```python
# Obtain a subset of data with nonmising values
nan_data = df.replace(to_replace = "?" , value = np.nan)
#nan_data.isin(['?']).any()
#nan_data.isna().any()
nonmissing_df = nan_data.dropna()
#nonmissing_df.isna().any()
nonmissing_df = nonmissing_df[10:20]
nonmissing_df
```

```python
# Fill missing values using any technique (mentioned in the lecture) according to data type
#nan_data.isnull().sum().sum()
#nan_data["workclass"].mode()[0]
nan_data["workclass"].fillna(nan_data["workclass"].mode()[0],inplace = True)
nan_data["occupation"].fillna(nan_data["occupation"].mode()[0],inplace = True)
nan_data["native-country"].fillna(nan_data["native-country"].mode()[0],inplace = True)
nan_data
```

#### Visualizing Data

1.Univariate analysis

```python
# "Age" distibution using histogram, pdf and cdf (using matplotlib  and seaborn)
%matplotlib inline
sns.displot(nan_data, x="age")
plt.show();
```

```python
counts,binedges = np.histogram(nan_data['age'],bins=10,density = True)

plt.xlabel('Age')
pdf = counts/(sum(counts))
cdf = np.cumsum(pdf)
plt.plot(binedges[1:],pdf);
plt.plot(binedges[1:],cdf);
```

```python
# Count plot for categorical attibute "education","martial-status" and "occupation" (using seaborn)
sns.countplot(y ='education', data = nan_data) 
plt.show()
```

3.Multivariate analysis

```python
# Calculate correlation of features and show using heatmap of Seaborn. 
corr = nan_data.corr()
sns.heatmap(corr,annot=True)
```

Heatmap provides check correlations among features and visualizing the correlation matrix. White fields have the most corelation and darkest fields least corelation among the features. Because of this reason the fields that have overlapping features are white. Looking at the heatmap in general, it is seen that the fnlwgt feature with the darkest areas is the least associated with the other features. The lightest color, that is the ones with the highest values are seen for hours per week. 
We can also examine the significance of the  features one by one, for example educational-num has the highest relationship with hours-per-week while it has the least relationship with fnlwgt.This makes our previous assessment of these two properties correct.

```python
# Draw seaborn pair plot demonstrating class info.
sns.pairplot(nan_data,height=3)
plt.show()
```

Pairs plot allows us to see the relationships between two variables. Histogram graph appears in features that conflict with it. The histogram allows us to see thepdf of a single variable. Other ones shows us the scatter plot. The scatter plots show us the relationship between the features. In general, it is seen that the acapital-gain feature is gathered at points close to zero as can be seen in all other features. This shows the interpretation that profit amounts are low. If we look at the capital gain and age features profit amount is distributed proportionally for all ages at the lowest and highest points. But if we look at the people who age higher than 84 and lower than 25, the profits appear to never occur at the top, indicating that the elder and young people are gaining low profits.
